```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
import pickle
```

```python
imgRoot = '../../data/coco_minitrain_25k/images_pruned/giraffes_elephants'
maskRoot = '../../data/coco_minitrain_25k/annotations'

# os.listdir(maskRoot)

with open(maskRoot+"/final_elephant_annotations.pickle", 'rb') as handle:
    eleAnnot = pickle.load(handle)

with open(maskRoot+"/final_giraffe_annotations.pickle", 'rb') as handle:
    girAnnot = pickle.load(handle)
```

```python
fig, ax = plt.subplots(10,3,figsize=(16,32))
for i, elephant in enumerate(list(eleAnnot.keys())[20:30]):
    img = plt.imread(imgRoot+"/elephant/"+elephant)
    mask = eleAnnot[elephant]['mask']
    ax[i,0].imshow(img)
    ax[i,1].imshow(img * mask[:, :, np.newaxis])
    ax[i,2].imshow(mask)
    
    ax[i,1].set_title(elephant)
plt.show()
```
